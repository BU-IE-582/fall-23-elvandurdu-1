##2. Moving Window Correlation

The 15 minute frequency data is converted into monthly stock prices in order to calculate correlation over a moving window.

```r
monthly_stock_prices <- to.period(dataw.xts.imputed, period = "months", OHLC = FALSE)
```

The stock pairs that I have chosen are Akbank(AKBNK) and Garanti (GARAN)​.

```r
correlation <- cor(monthly_stock_prices$AKBNK, monthly_stock_prices$GARAN)
```

```r
correlation
```

        GARAN
AKBNK 0.9446773

Then by using zoo library, I created a zoo object and calculated the 3 months moving window correlation between AKBNK and GARAN.

```r
monthly.stock.prices.zoo<-zoo(monthly_stock_prices[,c("AKBNK","GARAN")])
```

```r
rolling_correlation <- rollapply(monthly.stock.prices.zoo, width = 3, FUN = function(x) cor(x[,1], x[,2]), by.column = FALSE, align = "right") # 3 months moving correlation
rolling_correlation
```

```r
correlation_data <- data.frame(
  Date = index(rolling_correlation), 
  Correlation = coredata(rolling_correlation)  )
ggplot(correlation_data, aes(x = Date, y = Correlation)) +
  geom_line() +
  labs(title = "Moving Window Correlation Over Time", x = "Date", y = "Correlation") +
  theme_minimal()
```

 Since the stocks AKBNK and GARAN operate in the same sector, it is expected that they often exhibit similar price movements. It is due to shared market conditions affecting this particular sector. Moving correlation values are mostly high and this is consistent with the expected situation. High correlation values mean that they show similar movements. As it is seen in the graph, the correlation value significantly drops from 1 , especially at 3 points. Besides these points, the correlation coefficient is quite close to 1.
